```python
# os vetores MFCC podem variar de tamanho de acordo com a duração dos samples
# para uma CNN é necessário tamanhos de entrada fixos.
# então, é necessário colocar todos no msm tamanho, de acordo com o maior sample

import numpy as np

# define o tamanho max do vetor
max_pad_len = 174

def extract_features(file_name):
   
    try:
        # capta o audio vetorizado e seu sample-rate
        audio, sample_rate = librosa.load(file_name, res_type='kaiser_fast') 
        
        # extrai o vetor mfcc do audio
        mfccs = librosa.feature.mfcc(y=audio, sr=sample_rate, n_mfcc=40)
        
        # mostra o tamanho do mfcc (é bom pra verificar o maior item no vetor)
        print(mfccs.shape)
        
        # preenche com 0's os vetores menores
        pad_width = max_pad_len - mfccs.shape[1]
        mfccs = np.pad(mfccs, pad_width=((0, 0), (0, pad_width)), mode='constant')
        
    except Exception as e:
        print("Error encountered while parsing file: ", file_name)
        return None 
     
    return mfccs
```

```python
# Load various imports 
import pandas as pd
import os
import librosa

# path do dataset
fulldatasetpath = './audio/'

# path do .csv contendo as classes de cada arquivo do dataset
metadata = pd.read_csv('./metadata/UrbanSound8K.csv')

features = []

# itera por cada audio para extrair suas features
for index, row in metadata.iterrows():
    
    # cria o dir do audio a ser processado
    #file_name = os.path.join(os.path.abspath(fulldatasetpath),str(row["slice_file_name"]))
    file_name = os.path.join(os.path.abspath(fulldatasetpath),'fold'+str(row["fold"])+'/',str(row["slice_file_name"]))
    
    # extrai a classe daquele audio
    #class_label = row["class_name"]
    class_label = row["class"]
    
    # extrai as caracteristicas do audio
    data = extract_features(file_name)
    
    # adiciona no vetor de features
    features.append([data, class_label])

# converte o vetor de features em um dataFrame
featuresdf = pd.DataFrame(features, columns=['feature','class_label'])
print('Finished feature extraction from ', len(featuresdf), ' files') 

# armazena os dados para serem processados posteriormente
featuresdf.to_pickle('./data_processed/featuresData')  
print('Data saved sucessfull!')

# carrega os dados armazenados
df = pd.read_pickle('./data_processed/featuresData')
print(df)
```
